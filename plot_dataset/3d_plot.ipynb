```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balloon_data = np.random.normal(loc=0, scale=1, size=(100, 3))

melon_data = np.random.normal(loc=3, scale=1, size=(100, 3))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(balloon_data[:,0], balloon_data[:,1], balloon_data[:,2], c='red', label='Balloon')

ax.scatter(melon_data[:,0], melon_data[:,1], melon_data[:,2], c='green', label='Melon')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
ax.legend()
plt.show()
```
