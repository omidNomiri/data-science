```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("input/Covid-19_global_April.csv")
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"])

data["YearMonth"] = data["Date"].map(lambda date: 100*date.year + date.month)

data["new cases"] = data["New cases"]
new_case_data = data.groupby(["Country/Region"])["new cases"].sum().reset_index()
new_case_data = new_case_data.sort_values(by="new cases", ascending=False)
new_case_data.head(8)
```
